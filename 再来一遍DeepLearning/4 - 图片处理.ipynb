## 数字图像

#### 基础
    像素(Pixel): 是图像中的一个最小单位
    位图(bitmap): 通过记录每一个像素值来存储和表达的图像
    色彩深度/位深度： 位图中每个像素点要有多少个 二进制位 来表示
    BMP格式: windows系统的标准位图格式
    
- 二值位图(Binary Image)
    
    每个像素只有2种可能的取值，使用1位二进制来表示，位深度为1
        例子： 存储512 x 512像素的图像
        512x512bit=256kb=32KByte
- 灰度图像(Gray Image)
    
    每个像素使用1个字节表示，位深度为8，可以标识256中级别的灰度
        0标识黑色； 255标识白色
        例子：存储512x512像素的图像
        512 x 512bit = 256KByte
- 彩色图像(RGB)
    
    1. 每个像素都有红(R),绿(G),蓝(B)三个分量
    2. 一个像素点使用3个字节，位深度为24位
    3. 可以标识256x256x256=16777216种颜色
    4. 24位真彩色
    
|R|G|B|颜色|
|---|---|---|---|
|0|0|0|黑色|
|255|255|255|白色|
|255|0|0|红色|

- RGBA图像
    
    就是RGB图像 + 8位透明度信息Alpha
    ![4-Alpha参数引起图像变化.png](./img/4-Alpha参数引起图像变化.png)
    32位真彩色，1个像素点使用4个字节，位深度位32位
    
- 图像格式：
    - JEPG: 有损压缩，压缩率高，占空间小。适合色彩丰富、细节清晰细腻的大图像；不适合所按颜色较少，有大块颜色相近的图像。每次编辑都会降低图像质量，不适合多次编辑。
    - PNG: 无损压缩，适合有规律渐变色彩的图像
    - GIF: 支持 **静态格式** 和 **动态格式**。动态图片由*多幅图片*保存为一个图片，循环显示，形成动画效果。只能支持**256色**，适用于颜色简单，较小的小图像。
    - TIFF: 定义了四种不同格式
        - 二值图像
        - 黑白灰都图像
        - 带调色板的彩色图像
        - RGB真彩图像
        

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
img = Image.open('./lena.tiff')
img.save('lena.jpg')
img.save('lena.bmp')
```

```python
img1 = Image.open('lena.jpg')
img2 = Image.open('lena.bmp')
print(img1.format,img2.format)
```

### imshow()
plt.imshow(image对象/Numpy数组)

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
plt.axis('off')
plt.imshow(img)
plt.title(img.format)

plt.subplot(132)
plt.axis('off')
plt.imshow(img1)
plt.title(img1.format)

plt.subplot(133)
plt.axis('off')
plt.imshow(img2)
plt.title(img2.format)

plt.show()
```
