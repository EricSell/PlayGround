```python
import tensorflow as tf
```

```python
# 下载数据
boston_housing = tf.keras.datasets.boston_housing
# 分配数据
# load_data也有 test_split 默认 0.2 
(train_x,train_y),(test_x,test_y) = boston_housing.load_data()
```

```python
type(train_x)
print(train_x.ndim)
print(train_x.shape)
```

### 平均房间数与房价之间的关系 散点图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.family'] = 'SimHei'

plt.figure(figsize=(5,5))
#         平均房间数
plt.scatter(train_x[:,5],train_y)
plt.xlabel('平均房间数')
plt.ylabel('房价')
plt.title('平均房间数与房价之间的关系')
plt.show()
```

```python
titles = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE',
          'DIS','RAD','TAX','PTRATIO','B-1000','LSTAAT','MEDV']
plt.figure(figsize=(14,14))

for i in range(13):
    plt.subplot(4,4,(i+1))
    plt.scatter(train_x[:,i],train_y)
    plt.xlabel(titles[i])
    plt.ylabel('房价')
    
    plt.title(str(i)+'.'+titles[i]+' - Price')
plt.tight_layout()
plt.suptitle('各个属性与房价之间的关系',x=0.5,y=1.02,fontsize=20)
plt.show()
```
