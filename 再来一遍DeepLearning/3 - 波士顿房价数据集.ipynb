### Keras API


```python
import tensorflow as tf
```

```python
# 
boston_housing = tf.keras.datasets.boston_housing
# 
# load_data test_split  0.2 
(train_x,train_y),(test_x,test_y) = boston_housing.load_data()
```

```python
type(train_x)
print(train_x.ndim)
print(train_x.shape)
```

###  

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.family'] = 'SimHei'

plt.figure(figsize=(5,5))
#         
plt.scatter(train_x[:,5],train_y)
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```

```python
titles = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE',
          'DIS','RAD','TAX','PTRATIO','B-1000','LSTAAT','MEDV']
plt.figure(figsize=(14,14))

for i in range(13):
    plt.subplot(4,4,(i+1))
    plt.scatter(train_x[:,i],train_y)
    plt.xlabel(titles[i])
    plt.ylabel('')
    
    plt.title(str(i)+'.'+titles[i]+' - Price')
plt.tight_layout()
plt.suptitle('',x=0.5,y=1.02,fontsize=20)
plt.show()
```
