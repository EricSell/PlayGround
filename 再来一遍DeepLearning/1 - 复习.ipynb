##  Figure 对象
- num： 图形编号或名称，取值为数字/字符串
- figsize: 图像宽高（英寸）
- dpi：分辨率，缺省值为80
- facecolor：背景颜色
- edgecolor：边框颜色
- frameon：是否显示边框

```python
# 创建一个对象
plt.figure(figsize=(3,2),facecolor='green')
# 开始绘图
plt.plot()
plt.show()
```

### 散点图(Scatter) ： 数据点在直角坐标系统的分布图
scatter(x, y, scale, color, marker, label)
- x，y
- scale   数据点的大小
- color   数据点颜色
- marker   数据点样式
- label   图例文字


```python
import numpy as np
```

```python
# 标准正态分布
plt.rcParams['font.sans-serif'] = 'SimHei'  # 设置中文黑体
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号

n = 1024                                    # 1024个随机点
x1 = np.random.normal(0,1,n)                 # 生成 x坐标
y1 = np.random.normal(0,1,n)                 # 生成 y坐标
# 增加均匀分布的点
x2 = np.random.uniform(-4,4,(1,n))
y2 = np.random.uniform(-4,4,(1,n))



plt.scatter(x1,y1,color='blue',marker='*',label='正态分布')    # 绘制标准正态分布
plt.scatter(x2,y2,color='yellow',marker='o',label='均匀分布')  # 绘制均匀分布
plt.legend( )

plt.title('标准正态分布与均匀分布',fontsize=15) # 设置全局标题
# plt.text(2.5,2.5,'均 值：0\n标准差：1')     # 显示图内文本

plt.xlim(-4,4)                              # 设置x轴范围
plt.ylim(-4,4)                              # 设置y轴范围

plt.show()
```

## 折线图(Plot): 表述变量变化的趋势
plot(x,y,color,marker,label,linewidth,markersize)
- x    数据点的x坐标   0,1,2...
- y    数据点的y坐标 不可省略
- color 数据点的颜色
- marker 数据点的样式 
- label  图例文字
- linewidth  折现宽度
- markersize  数据点大小

```python
plt.rcParams['font.sans-serif'] = 'SimHei'

# 生成随机数列
n = 24
y1 = np.random.randint(27,37,n)
y2 = np.random.randint(40,60,n)

plt.plot(y1, label='温度')
plt.plot(y2, label='湿度')

plt.xlim(0,23)
plt.ylim(20,70)
plt.xlabel('小时',fontsize=12)
plt.ylabel('测量值',fontsize=12)

plt.title('24小时温度湿度统计',fontsize=16)

plt.legend()
plt.show()
```

## 柱形图(bar chart)： 用于数据的统计和分析
Bar(left, right, width, facecolor, edgecolor, label)

```python
y1 = np.random.randint(0,40,12)
y2 = np.random.randint(-40,0,12)

plt.bar(range(len(y1)),y1,width=0.8,facecolor='green',label='统计量1')
plt.bar(range(len(y2)),y2,width=0.8,facecolor='red',label='统计量2')

plt.title('柱状图',fontsize=20)

plt.legend( )
plt.show()
```
